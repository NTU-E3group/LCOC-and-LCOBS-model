```python
reload(lp)
df_vkt = pd.read_csv('./province_vkt.csv')
lower_case_vkt  = df_vkt["vkt_lower"]
upper_case_vkt  = df_vkt["vkt_upper"]

df_w_electiricity = pd.read_csv('./province_electiricity.csv')
lower_case_w_e = df_w_electiricity['c_w_e_lower']
upper_case_w_e = df_w_electiricity['c_w_e_upper']

sensitivity_result = pd.DataFrame({'type':['Compound',''],
                                   'lower bound':[lower_lcoc_calculator.cal_national_w_lcoc(),baseline_lcoc_calculator.cal_national_w_lcoc()],
                                   'upper bound':[upper_lcoc_calculator.cal_national_w_lcoc(),0]}) 

functions = [
    ([('setup_for_sen_charging_mix', (1,[1,0,0])), ('setup_for_sen_charging_mix', (0.6,[0,0,1]))],"Charging mix"),
    ([('setup_p_om_factor', (0.01,)), ('setup_p_om_factor', (0.2,))],"Public EVSE O&M cost"),
    ([('setup_vkt',(lower_case_vkt,)),('setup_vkt',(upper_case_vkt,))],"Vehicle kilometer traveled (VKT)"),
    ([('setup_FE', (0.20,)), ('setup_FE', (0.080,))],"EV fuel consumption rate"),
    ([('setup_provnice_c_w_electricity', (lower_case_w_e,)), ('setup_provnice_c_w_electricity', (upper_case_w_e ,))],"TOU: Workplace/Public"),
    ([('setup_c_r_capital', (810, 500)), ('setup_c_r_capital', (3349, 1000))],"Capital cost: L2 Res/Work"),
    ([('setup_c_p_capitial', ([15600, 22930, 33785, 79920],0.25)), ('setup_c_p_capitial', ([29250, 45000, 80000, 125000],0.75))],"Captial cost: Public"),
    ([('setup_L1_L2', (1, 0)), ('setup_L1_L2', (0, 1))],"Power level: Residential"),
    ([('setup_p_charging_mix', ([1, 0, 0, 0],)), ('setup_p_charging_mix', ([0,0,0,1],))],"Power level: Public")   
    # ([('setup_dr', (0.02,)), ('setup_dr', (0.07,))],"discount rate"),
    # ([('setup_evse_eff', (0.966,0.97755)), ('setup_evse_eff', (0.874,0.88445))],"EVSE eff") setup_provnice_c_w_electricity
]

# 使用迭代调用函数
for functions_to_call,name in functions:
    sen_lcoc_calculator = copy.deepcopy(baseline_lcoc_calculator)
    result = []
    for function_name, args in functions_to_call:
        method = getattr(sen_lcoc_calculator, function_name, None)
        if method and callable(method):
            method(*args)
        result.append(sen_lcoc_calculator.cal_national_w_lcoc())
    sensitivity_result = sensitivity_result.append({'type':name,'lower bound':result[0],'upper bound':result[1]}, ignore_index=True)

sensitivity_result
```
